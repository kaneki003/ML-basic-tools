# Seaborn KDE(kernal density estimation) plot

```python
drinks=pd.read_csv("https://bit.ly/drinksbycountry")
```

```python
drinks.head()
```

## Univariate 1D plot

```python
sns.kdeplot(drinks.beer_servings);
```

```python
sns.kdeplot(drinks.beer_servings,fill=True);
```

```python
sns.kdeplot(drinks.beer_servings,bw_adjust=2); #bandWidth by default chooses the one which suites the data
```

```python
sns.kdeplot(drinks.beer_servings,cumulative=True);
```

## Bivariate (2D) plot

```python
sns.kdeplot(data=(drinks.loc[:,['beer_servings','wine_servings']]),x='beer_servings',y='wine_servings');
```

```python
sns.kdeplot(data=(drinks.loc[:,['beer_servings','wine_servings']]),x='beer_servings',y='wine_servings',n_levels=20);
```

```python
sns.set_style('whitegrid')
sns.kdeplot(data=(drinks.loc[:,['beer_servings','wine_servings']]),x='beer_servings',y='wine_servings',fill=True);
```

```python
sns.kdeplot(data=(drinks.loc[:,['beer_servings','wine_servings']]),x='beer_servings',y='wine_servings',fill=True,cbar=True);
```

```python
import numpy as np
random_indices = np.random.choice(drinks.index, size=10, replace=False)
heat_df=drinks.loc[random_indices,:]
```

```python
heat_df['continent']
```

```python
data=heat_df.groupby('continent').total_litres_of_pure_alcohol.value_counts().unstack().fillna(0)
data
```

```python
sns.heatmap(data)
```

```python
data=np.array([[1,2],[3,4]])
```

```python
data.T #Transpose
```

```python
data=drinks[['beer_servings','wine_servings','spirit_servings','total_litres_of_pure_alcohol']].corr()
#Correlation Matrix
```

```python
sns.heatmap(data)
```

```python
#Color palette changes
```

```python
sns.heatmap(data, cmap="Blues")
```

```python
sns.heatmap(data, cmap="RdBu") 
#highlighting both high and low values
```
