# MatPlotLib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Simple plot of 4 numbers

```python
plt.plot([1,2,4,3])
plt.show()
```

# 2.Points with x and y values: Add title & axis labels

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.title('Parabolic_Curve',fontsize=18,color='b')
plt.xlabel('x')
plt.ylabel('x^2')
plt.show()
```

# 3.Change figure size,plot x's,set scales: x[0,6] y[0,20]

```python
plt.figure(figsize=(4,5))
plt.plot([1,2,3,4],[1,4,9,16] ,'kx')
plt.axis([0,6,0,20])
plt.annotate('same val',(1,1))
plt.show()
```

# Bar chart with four bars

```python
plt.clf() #clear figure
x=np.arange(4)
y=[8,8.5,3.2,5]
plt.xticks(x,('Ani','Laura','Ana','Chris'))
plt.bar(x,y, color=['y','r','k','g'])
plt.show()
```

# 5.Two sets of 10 random dots plotted

```python
d={ 'RedO':np.random.rand(10)
   ,'GrnX': np.random.rand(10)}
df=pd.DataFrame(d)
df.plot(style=['ro','gx'])
plt.show()
```

#  6.Time Series

```python
ts=pd.Series(np.random.randn(180), index=pd.date_range('1/1/2024',periods=180))
df=pd.DataFrame(np.random.randn(180,3),index=ts.index,columns=list('ABC'))
df.cumsum().plot()
plt.show()
```

## 7.Random Dots in Scatter

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
sizes=(30*np.random.rand(N))**2
plt.scatter(x,y,s=sizes,c=colors,alpha=0.5)
plt.show()
```

# 8.SubPlots

```python
fig=plt.figure()
fig.suptitle('SubPlots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222, facecolor='y')
fig.add_subplot(223)
fig.add_subplot(224)
plt.show()

# 221: topleft
# 222: topright
# 223: bottomleft
# 224: bottomright
```

# Saving figure into png

```python
fig=plt.figure()
fig.suptitle('SubPlots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222, facecolor='y')
fig.add_subplot(223)
fig.add_subplot(224)
plt.savefig('demo.png')  #Saving image
plt.show()
```
