# Duplicate Entries

```python
import pandas as pd
```

```python
ufo=pd.read_csv("https://bit.ly/uforeports")
```

```python
ufo.shape
```

```python
#Gives true if the entry above already exists
ufo.State.duplicated()
```

```python
#Total values which are duplicated
ufo.State.duplicated().sum()
```

```python
ufo.loc[ufo.duplicated(),:]
```

```python
#This required checking...
ufo.drop_duplicates(keep=False)
```
